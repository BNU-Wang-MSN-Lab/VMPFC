# Suppl. | Per-dataset split of single-motif predicted activation

For each motif, plot its two contributing image sets side-by-side (as `hue` grouping) to verify that the posterior → middle → anterior pattern in Fig. 2c is **not driven by a single dataset**. A consistent pattern across both datasets within a motif supports the generalizability claim.

| Motif | Datasets | Expected dominant region |
|---|---|---|
| social | PISC + Parade | anterior |
| valuation | Food_5k + Food_11 | middle |
| affect | ECED + GAPED | posterior |

**Activations are raw model units** (no z-score). Bars show across-image mean ± bootstrap 95% CI. The social panel uses fixed yticks (-0.015..0.030, step 0.005) to match the original figure.

## 1. Imports and global plot settings

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from utils import load_mean_predicted

plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.grid'] = False
PLOT_KWARGS = dict(dpi=300, transparent=True, bbox_inches='tight')
```

## 2. Helper functions: stack two datasets long-format, paired bar plot

```python
def stack_two_datasets(motif_folder, datasets):
    """Stack two datasets from one motif folder into long format for `hue` plotting.

    Each row: (Region, Activation, Dataset). `Region` cycles posterior/middle/anterior,
    one row per (image x region) per dataset.
    """
    pieces = []
    for ds in datasets:
        df = load_mean_predicted(motif_folder, ds)
        df_long = df.melt(var_name='Region', value_name='Activation')
        df_long['Dataset'] = ds
        pieces.append(df_long)
    return pd.concat(pieces, ignore_index=True)


def plot_paired(long_df, palette, save_stem, yticks=None):
    """Paired bar plot: 3 regions on x, two datasets as `hue`.

    Style: figsize=(4, 6), bootstrap 95% CI error bar (seaborn default),
    legend visible, xlabel + ylabel shown. Optional fixed yticks for the
    social panel (manuscript value range -0.015..0.030).
    """
    fig, ax = plt.subplots(figsize=(4, 6), dpi=100)
    sns.barplot(
        data=long_df,
        x='Region', y='Activation', hue='Dataset',
        order=['posterior', 'middle', 'anterior'],
        palette=palette, ax=ax,
    )
    ax.set_xlabel('vmPFC subregions')
    ax.set_ylabel('Predicted Activations')
    if yticks is not None:
        ax.set_yticks(yticks)
    # Dedupe legend entries (in case both bars share a label).
    handles, labels = ax.get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    ax.legend(by_label.values(), by_label.keys())
    fig.tight_layout()
    fig.savefig(f'../plots/{save_stem}.png', **PLOT_KWARGS)
    fig.savefig(f'../plots/{save_stem}.svg', **PLOT_KWARGS)
    plt.show()
```

## 3. social motif: PISC vs Parade

```python
# social motif: PISC (light blue) + Parade (dark blue).
# Fixed yticks match the original `plot_single.ipynb` social panel.
long_social = stack_two_datasets('social', ['PISC', 'Parade'])
plot_paired(
    long_social,
    palette=['#81B1D6', '#5182AF'],
    save_stem='split_social_PISC_Parade',
    yticks=np.arange(-0.015, 0.031, 0.005),
)
```

## 4. valuation motif: Food_5k vs Food_11

```python
# valuation motif: Food_5k (pale yellow) + Food_11 (bright yellow).
long_value = stack_two_datasets('value', ['Food_5k', 'Food_11'])
plot_paired(
    long_value,
    palette=['#eeeca2', '#dcd844'],
    save_stem='split_value_Food5k_Food11',
)
```

## 5. affect motif: ECED vs GAPED

```python
# affect motif: ECED_noneutral (pale red) + GAPED_negative_nonsocial (bright red).
long_affect = stack_two_datasets('emotion', ['ECED_noneutral', 'GAPED_negative_nonsocial'])
plot_paired(
    long_affect,
    palette=['#e08080', '#c00000'],
    save_stem='split_affect_ECED_GAPED',
)
```
