```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from tqdm import tqdm
from scipy.stats import ttest_ind, ttest_1samp, levene
import joblib
import numpy as np
```

```python
DATA_PATH = Path('../data')
RESULTS_PATH = Path('../results')

X_NAME_LIST = ['SC', 'FC']
Y_NAME_LIST = ['social', 'valuation', 'affect']
```

```python
for X_name in X_NAME_LIST:
    for y_name in Y_NAME_LIST:
        # load data
        df = joblib.load(DATA_PATH / f'inhouse_BNA/{X_name}_{y_name}_prediction.df.pkl')
        mean_r_value = df.iloc[0, :].mean()
        permutation_r_value = df.iloc[1:, :].mean(axis=1)
        p = (np.sum(mean_r_value < permutation_r_value) + 1) / (len(permutation_r_value) + 1)
        print(f'{X_name}_{y_name}: permutation p = {p}')
```
