```python
from pathlib import Path
import shutil
```

```python
to_folder = Path('../results/cbptools_output')
to_folder.mkdir(parents=True, exist_ok=True)
```

```python
for from_folder in Path('../results/cbptools').iterdir():
    this_to_folder = to_folder / from_folder.name
    this_to_folder.mkdir(parents=True, exist_ok=True)
    shutil.copyfile(
        from_folder / 'individual/internal_validity.tsv',
        this_to_folder / 'internal_validity.tsv')

    for i in range(2, 7):
        this_from_file = from_folder / f'group/{i}clusters/labeled_roi.nii.gz'
        if not this_from_file.exists():
            print(this_from_file)
        else:
            shutil.copyfile(
                this_from_file,
                this_to_folder / f'K{i}.nii.gz'
            )
```
