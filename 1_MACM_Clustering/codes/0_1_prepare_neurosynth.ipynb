# Introduction: Neurosynth Data Initialization

This notebook serves as the foundational data preparation step for our meta-analytic pipeline. Its primary objective is to initialize, construct, and serialize the core `Dataset` object required by the `neurosynth` Python library.

```python
from neurosynth import Dataset
from pathlib import Path
import tarfile
import warnings


warnings.filterwarnings("ignore", category=FutureWarning)

DATA_PATH = Path('../data')
PLOTS_PATH = Path('../plots')
RESULTS_PATH = Path('../results')

DATASET_FILE = DATA_PATH / 'neurosynth_data/dataset.pkl'
```

Data was downloaded from **[github: neurosynth/neurosynth-data](https://github.com/neurosynth/neurosynth-data/blob/e8f27c4a9a44dbfbc0750366166ad2ba34ac72d6)**. The dataset provided in this repository is sourced from Neurosynth and is licensed under the Open Database License (ODbL). Once downloaded, please place the file in the `../data/neurosynth_data/current_data.tar.gz`directory and run the cell below.

Due to the large file size, we are not sharing the `.txt` or `.pkl` files directly. However, here are the instructions on how to generate them. The code below simply processes the extracted data into a `.pkl` format, which allows for much faster loading times.

```python
if not DATASET_FILE.exists():
    with tarfile.open(DATA_PATH / 'neurosynth_data/current_data.tar.gz', 'r:gz') as tar:
        tar.extractall(path=DATA_PATH / 'neurosynth_data')
    dataset = Dataset(DATA_PATH / 'neurosynth_data/database.txt')
    dataset.add_features(str(DATA_PATH / 'neurosynth_data/features.txt'))
    dataset.save(DATASET_FILE)
else:
    dataset = Dataset.load(DATASET_FILE)
print(f'Dataset includes {dataset.feature_table.data.shape[0]} studies.')
```
