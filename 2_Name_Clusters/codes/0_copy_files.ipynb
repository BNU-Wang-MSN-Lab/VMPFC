# Data Initialization and Workspace Setup

This notebook serves as the foundational setup script for the current analysis pipeline. Because our workflow builds directly upon the data previously processed in the MACM (Meta-Analytic Connectivity Modeling) Clustering phase, this script automates the transfer of those foundational files into our local workspace.

Specifically, this script performs the following data management operations:
1. **Directory Initialization:** Establishes the standard project structure (`data`, `results`, `plots`, `documents`).
2. **Data Migration:** Systematically copies the curated Neurosynth dataset, term mapping dictionaries, and region-of-interest (ROI) masks from the `1_MACM_Clustering` parent directory into the local `data` folder.

This ensures that our current analysis is completely reproducible and operates on an isolated, localized copy of the source data.

```python
import pandas as pd
from pathlib import Path
import shutil


DATA_PATH = Path('../data')
RESULTS_PATH = Path('../results')
PLOTS_PATH = Path('../plots')
DOCUMENTS_PATH = Path('../documents')
```

```python
(DATA_PATH / 'neurosynth_data').mkdir(parents=True, exist_ok=True)

shutil.copyfile(
    '../../1_MACM_Clustering/data/neurosynth_data/term_mapping.json',
    DATA_PATH / 'neurosynth_data/term_mapping.json' )


shutil.copyfile(
    '../../1_MACM_Clustering/data/neurosynth_data/dataset.pkl',
    DATA_PATH / 'neurosynth_data/dataset.pkl', )
```

```python
shutil.copytree(
    '../../1_MACM_Clustering/data/masks',
    DATA_PATH / 'masks',
    dirs_exist_ok=True
)
```
