# Ext. Data Fig. 3b | Motif-intersection predicted activation

Each panel shows one image set that probes a **motif intersection** — e.g., CFD faces engage both social-cognition and valuation systems, NAPS_ERO erotica engages all three.

| Panel | Image set | Probes |
|---|---|---|
| 1 | CFD       | social ∩ value |
| 2 | Antique   | value ∩ affect |
| 3 | SMID      | social ∩ affect |
| 4 | NAPS_ERO  | social ∩ value ∩ affect |

Bars show across-image mean ± **bootstrap 95% CI** (seaborn default). Activations are the across-subject mean predicted response per image, in raw model units (no z-score).

**Statistics:** per-region one-sample t-test against 0 with FDR-BH correction across the 3 regions. Caption convention: asterisks denote `p_FDR < 0.05` significance above zero. Asterisks are not drawn here; add in PPT.

**Layout:** figsize=(3, 3) per panel (wider than the original (2,3) per request).

## 1. Imports and global plot settings

```python
# seaborn for default bootstrap CI error bars, matches original `plot_single.ipynb` cells.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Per-region one-sample t-test vs 0, FDR-BH correction across 3 regions.
from scipy.stats import ttest_1samp
from statsmodels.stats.multitest import multipletests

from utils import load_mean_predicted

plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.grid'] = False
PLOT_KWARGS = dict(dpi=300, transparent=True, bbox_inches='tight')

# Single grey hue for intersection bars (manuscript caption: "bars are gray for readability").
INTERSECTION_COLOR = '#9e9e9e'
```

## 2. Helper functions: per-region t-test vs 0 and the bar panel

```python
def stats_vs_zero(df_wide, panel_name):
    """Per-region one-sample t-test vs 0, with FDR-BH correction across 3 regions.

    `df_wide` is the across-subject mean DataFrame (one row per image, columns
    posterior/middle/anterior). We test whether each region's mean activation
    is significantly above (or below) zero. Output goes to stdout only.
    """
    print(f'=== {panel_name}: one-sample t-test vs 0, FDR-BH corrected ===')
    pvals = []
    for region in ['posterior', 'middle', 'anterior']:
        t, p = ttest_1samp(df_wide[region].values, 0.0)
        pvals.append(p)
        print(f'  {region:10s}: t = {t: .4f}, p_uncorrected = {p:.4g}')
    _, pvals_fdr, _, _ = multipletests(pvals, method='fdr_bh')
    for region, p in zip(['posterior', 'middle', 'anterior'], pvals_fdr):
        print(f'  {region:10s}: p_FDR = {p:.4g}')


def plot_panel(df_wide, save_stem):
    """Bar plot for one intersection panel.

    Layout: figsize=(3, 3) (wider than original (2,3) per user request),
    single grey color, default seaborn bootstrap 95% CI error bar.
    """
    plot_df = df_wide[['posterior', 'middle', 'anterior']]
    fig, ax = plt.subplots(figsize=(3, 3), dpi=100)
    sns.barplot(data=plot_df, color=INTERSECTION_COLOR, width=0.5, ax=ax)
    ax.set_xlabel('')
    ax.set_ylabel('')
    fig.tight_layout()
    fig.savefig(f'../plots/{save_stem}.png', **PLOT_KWARGS)
    fig.savefig(f'../plots/{save_stem}.svg', **PLOT_KWARGS)
    plt.show()
```

## social ∩ value (CFD)

```python
# Load the across-subject mean predicted CSV for `CFD` under `social_value/`.
df = load_mean_predicted('social_value', 'CFD')

# Per-region t-test (printed only, no plot annotation).
stats_vs_zero(df, 'social ∩ value (CFD)')

# Bar plot with bootstrap 95% CI.
plot_panel(df, 'inter_socialvalue_CFD')
```

## value ∩ affect (Antique)

```python
# Load the across-subject mean predicted CSV for `Antique` under `value_emotion/`.
df = load_mean_predicted('value_emotion', 'Antique')

# Per-region t-test (printed only, no plot annotation).
stats_vs_zero(df, 'value ∩ affect (Antique)')

# Bar plot with bootstrap 95% CI.
plot_panel(df, 'inter_valueaffect_Antique')
```
